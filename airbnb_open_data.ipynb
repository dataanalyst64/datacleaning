```python
import pandas as pd
```

```python
data=pd.read_csv("Airbnb_Open_Data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
#as we see that there are null values in different columns and we see that house rules and licence have no work for our client requirements so we delete it
data.drop(columns=['house_rules', 'license'],inplace=True)
```

```python
data.columns
```

```python
#now we convert the last review column type to date type
data['last review']=pd.to_datetime(data['last review'], errors = 'coerce')
```

```python
data.info()
```

```python
#now we fill null values of column "review per month" with 0
data['reviews per month'].fillna("0",inplace=True)
```

```python
data['reviews per month'].isna().sum()
```

```python
data.isna().sum()
```

```python
data['last review']=data['last review'].fillna(data['last review'].min())
```

```python
data['last review'].isna().sum()
```

```python
data.dropna(subset =['NAME', 'host name'],inplace=True)
```

```python
data['NAME'].isna().sum()
```

```python
data.head()
```

```python
data.to_csv("cleaned_airbnb_data.csv",index=False)
```
